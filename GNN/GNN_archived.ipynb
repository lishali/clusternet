```python
#online version of our variables:

signal_dim = 3
communities = 2 #number of communities, chance to 
group_size = 2
Size = 100
p_min = 0.5
p_max = 0.6


#A = balanced_stochastic_blockmodel(communities=communities, groupsize=group_size, p_in=0.6, p_out=0.01)
#Adj = tf.cast(A, dtype = tf.float32)
DATA = [np.asarray(balanced_stochastic_blockmodel(communities, group_size, p, 0.1*p)).astype(np.double) for p in np.linspace(p_min, p_max, Size)]

Adj = tf.placeholder(dtype=tf.float32, shape=[communities*group_size, communities*group_size])
#initialize a 10 dim signal
F = tf.Variable(tf.random_normal([communities*group_size, signal_dim], stddev=1.0, mean=0.0, seed=1))

#initialize some 10x10 weights

C_a = tf.Variable(tf.random_normal([signal_dim, signal_dim], stddev=1.0, mean=0.0))
C_a = tf.expand_dims(tf.transpose(C_a), 0)
C_b = tf.Variable(tf.random_normal([signal_dim, signal_dim], stddev=1.0, mean=0.0))
C_b = tf.expand_dims(tf.transpose(C_b), 0)

DF1 = tf.expand_dims(Diag(Adj, F), 2) #so we batch the matrix multiply correctly
WF1 = tf.expand_dims(Diffusion(Adj, F), 2)
 
A1 = tf.reduce_sum(tf.mul(C_a, DF1), -2) #-2 because you transposed C_a so that we can leverage broadasting in tf.mul to do our tf.matmul
B1 = tf.reduce_sum(tf.mul(C_b, WF1), -2)
Psi1 = A1 + B1



init = tf.initialize_all_variables()

with tf.Session() as sess:
    for i in xrange(10):
        sess.run(init)
        print sess.run([A1, B1, Psi1], feed_dict={Adj: DATA[i]})
```
