```python
import matplotlib.pyplot as plt

%matplotlib inline

import networkx as nx

G=nx.nx.planted_partition_graph(l=2, k=4, p_in=1, p_out =0.15)
pos=nx.shell_layout(G) # positions for all nodes

# nodes
nx.draw_networkx_nodes(G,pos,
                       nodelist=[0,1,2,3],
                       node_color='r',
                       node_size=500,
                   alpha=0.8)
nx.draw_networkx_nodes(G,pos,
                       nodelist=[4,5,6,7],
                       node_color='b',
                       node_size=500,
                   alpha=0.8)

# edges
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
nx.draw_networkx_edges(G,pos,
                       edgelist=[(0,1),(1,2),(2,3),(3,0), (2,0), (3,1)],
                       width=8,alpha=0.5,edge_color='r')
nx.draw_networkx_edges(G,pos,
                       edgelist=[(4,5),(5,6),(6,7),(7,4), (6,4), (7,5)],
                       width=8,alpha=0.5,edge_color='b')
# some math labels
labels={}
labels[0]=r'$a$'
labels[1]=r'$b$'
labels[2]=r'$c$'
labels[3]=r'$d$'
labels[4]=r'$\alpha$'
labels[5]=r'$\beta$'
labels[6]=r'$\gamma$'
labels[7]=r'$\delta$'
nx.draw_networkx_labels(G,pos,labels,font_size=16)

plt.axis('off')
plt.savefig("labels_and_colors.png") # save as png
#plt.show() 
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

import networkx as nx

G=nx.nx.planted_partition_graph(l=4, k=1, p_in=1, p_out =1.0)
pos=nx.circular_layout(G) # positions for all nodes

# nodes
nx.draw_networkx_nodes(G,pos,
                       nodelist=[0,1,2,3],
                       node_color='y',
                       node_size=1000,
                   alpha=0.8)


# edges
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
nx.draw_networkx_edges(G,pos,
                       edgelist=[(0,1),(1,2),(2,3),(3,0),(2,0),(3,1)],
                       width=8,alpha=0.5,edge_color='y')

# some math labels
labels={}
labels[0]=r'$w_{11}$'
labels[1]=r'$w_{22}$'
labels[2]=r'$w_{33}$'
labels[3]=r'$w_{44}$'

nx.draw_networkx_labels(G,pos,labels,font_size=16)

plt.axis('off')


plt.savefig("SBM.png") # save as png
plt.show() 
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

import networkx as nx

G=nx.nx.planted_partition_graph(l=2, k=20, p_in=0.5, p_out =0.1, seed=None)
pos_spectral=nx.spectral_layout(G) # positions for all nodes

# nodes
nx.draw_networkx_nodes(G,pos_spectral,
                       nodelist=[i for i in range(20)],
                       node_color='r',
                       node_size=200,
                   alpha=0.8)
nx.draw_networkx_nodes(G,pos_spectral,
                       nodelist=[i+20 for i in range(20)],
                       node_color='b',
                       node_size=200,
                   alpha=0.8)

nx.draw_networkx_edges(G,pos_spectral,width=1.0,alpha=0.5)

# some math labels

plt.axis('off')
plt.savefig("SBM_balanced_large_dense_spectral_4.png") # save as png
#plt.show() 

```
